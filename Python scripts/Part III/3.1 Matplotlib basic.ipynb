```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Datos
x = np.linspace(-10, 10, 500)
curvas = [
    {"media": 0, "desviacion": 1, "color": "blue"},
    {"media": 2, "desviacion": 2, "color": "green"},
    {"media": -3, "desviacion": 1.5, "color": "red"},
]

# Graficar
plt.figure(figsize=(10, 6))
for curva in curvas:
    y = norm.pdf(x, curva["media"], curva["desviacion"])
    plt.plot(x, y, label=f"={curva['media']}, ={curva['desviacion']}", color=curva["color"])

# Personalizacin
plt.title("Distribuciones Normales")
plt.xlabel("X")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos Iris
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Species"]
iris_data = pd.read_csv(url, header=None, names=column_names)

```

```python
# Crear histogramas por especie
species = iris_data["Species"].unique()

for specie in species:
    subset = iris_data[iris_data["Species"] == specie]
    plt.hist(subset["Sepal Length"], bins=10, alpha=0.6, label=specie, edgecolor='black')

plt.title('Comparacin de Longitud del Spalo por Especie')
plt.xlabel('Longitud del Spalo (cm)')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
