```python
#Regresion lineal polinomica 

#ejemplo 1 :
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#dataset
X = np.array([1 ,2 ,3 ,4,5]).reshape(-1,1)
y = np.array([6,11,18,27,38]).reshape(-1,1)

#creamos el modelo
modelo = LinearRegression()

#Entrenar el modelo
modelo.fit(X,y)

#Modelo predictivo 
y_predic = modelo.predict(X)

#Metricas 
mse = mean_squared_error(y , y_predic)
r2 = r2_score(y , y_predic)
print(r2)
print(mse)

#graficar
plt.scatter(X,y , color='red')
plt.plot(X , y_predic , color = 'green')
plt.show()
```

```python
#Regresion lineal polinomica 

#ejemplo 1 :
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#dataset
X = np.array([1 ,2 ,3 ,4,5]).reshape(-1,1)
y = np.array([6,11,18,27,38]).reshape(-1,1)

#creamos caracteristicas polinomicas
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

#Ajustamos el modelo
model = LinearRegression()

#entrenamos
model.fit(X_poly , y)

#predicciones
y_pred = model.predict(X_poly)

#imprimir metricas sin definicion
print(f"R2 Score {r2:.2f}")#R2
print(f"error cuadratico medio {mse:.2f}")#MSE

#graficar
plt.scatter(X,y , color='red')
plt.plot(X , y_pred , color = 'green')
plt.title('Regresion lineal Polinomica')
plt.show()
```
