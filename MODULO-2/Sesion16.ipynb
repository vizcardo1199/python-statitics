```python
#Regresion lineal polinomica 

#ejemplo 1 :
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#dataset
X = np.array([1 ,2 ,3 ,4,5]).reshape(-1,1)
y = np.array([6,11,18,27,38]).reshape(-1,1)

#creamos el modelo
modelo = LinearRegression()

#Entrenar el modelo
modelo.fit(X,y)

#Modelo predictivo 
y_predic = modelo.predict(X)

#Metricas 
mse = mean_squared_error(y , y_predic)
r2 = r2_score(y , y_predic)
print(r2)
print(mse)

#graficar
plt.scatter(X,y , color='red')
plt.plot(X , y_predic , color = 'green')
plt.show()
```

```python
#Regresion lineal polinomica 

#ejemplo 1 :
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#dataset
X = np.array([1 ,2 ,3 ,4,5]).reshape(-1,1)
y = np.array([6,11,18,27,38]).reshape(-1,1)

#creamos caracteristicas polinomicas
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

#Ajustamos el modelo
model = LinearRegression()

#entrenamos
model.fit(X_poly , y)

#predicciones
y_pred = model.predict(X_poly)

#imprimir metricas sin definicion
print(f"R2 Score {r2:.2f}")#R2
print(f"error cuadratico medio {mse:.2f}")#MSE

#graficar
plt.scatter(X,y , color='red')
plt.plot(X , y_pred , color = 'green')
plt.title('Regresion lineal Polinomica')
plt.show()
```

```python
#usando excel para regresion lineal

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


#leemos el archivo del excel
direccion_del_archivo = 'regresion-lineal1.csv'
dataset = pd.read_csv(direccion_del_archivo) # esto solo funciona para archivos con extensiones.csv
#para excel o xlsx pd.read_excel()

#extraer las variables X & Y
X = dataset[['X']].values # con esto tenemos la columna de X
y = dataset[['y']].values # con esto tenemos la columna de Y

#aplicamos regreslion lineal
modelo = LinearRegression()
modelo.fit(X,y)
y_pred = modelo.predict(X)

#plotearlo 
plt.scatter(X ,y , color ='blue')
plt.plot(X , y_pred , color='purple')
plt.show()
```

```python
#usando excel para regresion lineal multiple

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


#leemos el archivo del excel
direccion_del_archivo = 'rl-mul0.xlsx'
df = pd.read_excel(direccion_del_archivo) # esto solo funciona para archivos con extensiones.csv
#para excel o xlsx pd.read_excel()

#preparemos los datos 
X = df[['Publicidad','Poblacion','Ingreso', 'Desempleo']]
y = df['Ventas']

#dividiendo datos en conjunto de entrenamiento y prueba
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.2) # entrenar 80%, 20%test ( 0-1)
#esta funciona se puede aplicar para todos los metodos de regresion lineal

#crear el modelo 
modelo = LinearRegression()
modelo.fit(X_train , y_train)

#modelo predictivo
y_pred = modelo.predict(X_test)

#metrica
mse = mean_squared_error(y_test ,y_pred ) # mse(los datos , el modelo)
print(mse)


#probemos el modelo 
#vamos una predicción publicidad 45k , población 1 , ingreso 35000 , desempleo 4
ventas1 = np.array([[45000 , 1 , 35000 , 4]])
prediccion1 = modelo.predict(ventas1)
print(prediccion1 )


```

```python
# correlaciónes

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#crear el DataFrame 
data = pd.DataFrame({
    'X1':[230 , 44 , 17 , 151 , 180], #PUBLICIDAD TV
    'X2':[37 , 39 , 45 , 41 , 10],#PUBLICIDAD RADIO
    'X3':[69 , 45 , 58 , 89 , 58],#PUBLICIDAD ONLINE
    'X4':[5 , 7 , 3 , 2 , 9],#DESCUENTOS
    'X5':[10 , 15 , 20 , 25 , 12],#COMPETENCIAS
    'Y': [22 , 10 , 7 , 18 , 20]#VENTAS
})

#correlacion de pearson
matriz_correlacion = data.corr()

#filtramos las correlaciones con Y
correlacion_con_y = matriz_correlacion['Y'].drop('Y')

#Ordenamos las correlaciones por magnitud
correlación_ordenada =correlacion_con_y.abs().sort_values(ascending=False) 

print('Correlaciones con Y')
print(correlación_ordenada)

#mapa de calor de correlaciónes
plt.figure(figsize=(10,6))
sns.heatmap(matriz_correlacion , annot=True , cmap='coolwarm' , fmt=".2f")
plt.title('Mapa de calor de correlaciones')
plt.show()
```
