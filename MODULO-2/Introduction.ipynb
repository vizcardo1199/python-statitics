```python
# Circular O Pie(plt.Pie) , Lineas(plt.plot) , barras vertical (plt.bar) , Dispersion(plt.scatter)
# ,barra horizontal(plt.barh)
#Histograma (plt.hist)
# Datos reales: edades de un grupo de personas

import matplotlib.pyplot as plt
data = [22, 25, 28, 22, 23, 24, 30, 35, 30, 40, 42, 45, 50, 28, 25, 22, 35, 40, 50, 55, 60, 65, 70, 75, 80, 85]


#CREAR HISTOGRAMA 
plt.figure() # sirve para otorgarle un framework y un tamaño
plt.hist(data , bins=5 , edgecolor='red' , alpha=0.7)
plt.title('Histograma de edades')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Datos reales: edades de un grupo de personas categorizadas por género
edades_hombres = [22, 25, 28, 22, 23, 24, 30, 35, 30, 40, 42, 45, 50, 28, 25, 22, 35, 40, 50, 55, 60, 65, 70, 75]
edades_mujeres = [23, 26, 29, 24, 25, 26, 31, 36, 32, 41, 43, 46, 51, 30, 27, 24, 37, 41, 53, 58, 62, 66, 72, 78]

plt.hist(
    [edades_hombres,edades_mujeres], #datos de hombres y mujeres
    bins = 10 , # dividir en 10 rangos (bins)
    edgecolor = 'green' ,# Bordes de color 'green'
    alpha = 0.8 ,# transparencia
    label = ['Hombres','Mujeres'],
    color = ['green','pink'], # metodo para dar colores al histograma
    stacked = True
)
plt.legend()
plt.show()
```

```python
# pip install matplotlib
#DIAGRAMA DE VIOLIN 
 
import matplotlib.pyplot as plt
import seaborn as sns

# Datos reales: edades de un grupo de personas categorizadas por género
edades_hombres = [22, 25, 28, 22, 23, 24, 30, 35, 30, 40, 42, 45, 50, 28, 25, 22, 35, 40, 50, 55, 60, 65, 70, 75]
edades_mujeres = [23, 26, 29, 24, 25, 26, 31, 36, 32, 41, 43, 46, 51, 30, 27, 24, 37, 41, 53, 58, 62, 66, 72, 78]

#creacion del diagrama de violin 
sns.violinplot(data = [edades_hombres ,edades_mujeres ] , palette=['skyblue', 'salmon'] , split=False)
plt.xticks([0,1] , ['Hombres' , 'Mujeres'])
plt.show()

#Mas ancho  => mayor densidad de datos en ese rango
#Mas estrecho => menor densidad de datos en ese rango
```

```python
# FACEGRIT PARA HISTOGRAMAS DE INGRESOS POR GENERO
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np  # np.array 

#datos simulados 
data = pd.DataFrame({
    'Edad':np.random.randint(18,60,100),# (desde, hasta , cuantos)
    'Genero': np.random.choice(['Hombre','Mujer'] , 100), # ([data a elegir] , cuantos)
    'Ingreso': np.random.randint(2000 , 8000 , 100)
})

#Facet grid para histogramas de ingresos por genero
g=sns.FacetGrid(data , col='Genero' , height = 4 , aspect=1.2)
# height : define la altura del subgrafico , aspect : define la relacion ancho-altura para los graficos
g.map(sns.histplot , 'Ingreso' ,kde = True , color = 'red' )
# kde: Superte una curva de densidad al histograma mostrandooslas una representación mas suavizada 
# de la distribución
g.set_titles("{col_name}")
plt.show()

```
