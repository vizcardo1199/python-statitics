```python
import numpy as np
#graficar multiples funciones 
# y1 = x^2 , y2 = x^3 , y3 = sen(x)
x = np.linspace(-10 , 10 ,500) # 500 puntos entre [-10 y 10]
y1 = (x**2)*10
y2 = x**3
y3 = (np.sin(x))*100

plt.figure(figsize=(12,8)) # ajusta el tamańo del grafico
plt.plot(x,y1,label='Funcion cuadratica' )
plt.plot(x,y2,label='Funcion cubica')
plt.plot(x,y3,label='Funcion seno')

#personalizar el grafico
plt.title('Grafico de multiples funciones',fontsize=10)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid()
plt.show()
```

```python
#multiples curvas de distribución normal 
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

#Datos
x = np.linspace(-10,10,500)

#diccionario o un objeto de datos de las curvas

curvas=[
    {"media":0 , 'desviacion':1 , 'color': 'blue'},
    {"media":2 , 'desviacion':2 , 'color': 'green'},
    {"media":-3 , 'desviacion':1.5 , 'color': 'red'}
]
plt.figure(figsize=(10,6))
#y1 = ,y2= 

for curva in curvas:
    y = norm.pdf(x,curva['media'],curva['desviacion'])
    plt.plot(x,y,label=curva['media'],color=curva["color"])

plt.legend()
plt.show()






```

```python
#GRAFICO DE BARRAS AGRUPADOS
import numpy as np
import matplotlib.pyplot as plt

#datos inventados
#ejemplos 
categorias = ['A','B','C','D','E']# jalas las categorías de bd
valores = [10,20,30,40,50]
# barras agrupadas 
subcategorias=['X','Y']# NUMERO DE SUBCATEGORIAS DENTRO DE CADA CATEGORIA
valores_agrupados = [ [10,15],[20,25],[15,10],[25,20],[30,35]] # BIDIMENSIONAL 


x = np.arange(len(categorias))# posiciones del eje x basado en categorias
ancho = 0.3

plt.figure(figsize=(10,8)) #AJUSTA EL TAMAŃO DEL GRAFICO
#algoritmo de recorrer las categorias con las subcategorias
for i , subcat in enumerate(subcategorias): # dibujas las barras para cada subcategoria
    plt.bar(x + i*ancho, [v[i] for v in valores_agrupados],width = ancho ) # comprehension de funciones

plt.title('Grafico de barras bidimensionales')
plt.show()




```

```python
# grafico de barras con lineas de referencia o cota inferior y cota superior
import numpy as np
import matplotlib.pyplot as plt

#datos inventados
#ejemplos 
categorias = ['A','B','C','D','E']# jalas las categorías de bd
valores = [10,20,30,40,50]

plt.bar(categorias, valores)
plt.axhline(y=20 , color='green',label='Limite inferior' , linestyle='--')# esto me hace las lineas punteadas
plt.axhline(y=40 , color='red',label='Limite superior')
plt.grid(alpha=0.2) # alpha=[0-1] me da una transparencia
plt.title('G. Barras con lineas de referencia')
plt.show()
```

```python
# Graficos 3d
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Superficie basica
x= np.linspace(-5,5,50)
y= np.linspace(-5,5,50)

X,Y= np.meshgrid(x,y)#crear el plano

Z= np.sin(np.sqrt(X**2+Y**2))# es la funcion que se eleva en el plano

#este es el algoritmo para crear el dibujo 3d
fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X , Y , Z ,cmap='viridis')
plt.show()





```
