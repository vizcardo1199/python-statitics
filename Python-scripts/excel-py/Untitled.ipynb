```python
import pandas as pd

# Carga el archivo Excel
data = pd.ExcelFile("datos_ejemplo.xlsx")

# Carga cada hoja
ventas_productos = data.parse("Ventas_Productos")
temperaturas_ciudades = data.parse("Temperaturas_Ciudades")
poblacion_paises = data.parse("Población_Países")

import matplotlib.pyplot as plt

# Ejemplo de gráfico de barras
productos = ventas_productos['Producto'].unique()
cantidad_vendida = ventas_productos.groupby('Producto')['Cantidad_Vendida'].sum()

plt.bar(productos, cantidad_vendida)
plt.title('Cantidad Vendida por Producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.show()
```
