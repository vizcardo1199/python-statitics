```python
import pandas as pd
import matplotlib.pyplot as plt

#cargar el archivo excel
data = pd.ExcelFile('datos_ejemplo.xlsx')

#cargamos las hojas
ventas_productos = data.parse('Ventas_Productos') # elarchivoexcelcargado.parse('el nombrede la hoja)
temperaturas_ciudades = data.parse('Temperaturas_Ciudades')
poblacion_paises = data.parse('Población_Países')


#GRAFICAR POBLACION PAISES
```

```python
#grafico de barras 
productos = ventas_productos['Producto'].unique()
cantidad_vendida = ventas_productos.groupby('Producto')['Cantidad_Vendida'].sum()

plt.bar(productos,cantidad_vendida)
plt.show()
```

```python
# Grafico de lineas (VENTAS POR PRODUCTO ESPECIFICO)
#filtrar los datos del producto A

producto_a = ventas_productos[ventas_productos['Producto']== 'Producto A']
fechas = producto_a['Fecha']

plt.figure(figsize=(12,4)) # con esto le dimos tamaño a la figura
plt.plot(fechas , producto_a['Cantidad_Vendida'] , marker='o')
plt.show()
```
