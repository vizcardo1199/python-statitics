```python
import pandas as pd
import matplotlib.pyplot as plt

#cargar el archivo excel
data = pd.ExcelFile('datos_ejemplo.xlsx')

#cargamos las hojas
ventas_productos = data.parse('Ventas_Productos') # elarchivoexcelcargado.parse('el nombrede la hoja)
temperaturas_ciudades = data.parse('Temperaturas_Ciudades')
poblacion_paises = data.parse('Población_Países')


#GRAFICAR POBLACION PAISES
```

```python
#grafico de barras 
productos = ventas_productos['Producto'].unique()
cantidad_vendida = ventas_productos.groupby('Producto')['Cantidad_Vendida'].sum()

plt.bar(productos,cantidad_vendida)
plt.show()
```

```python
# Grafico de lineas (VENTAS POR PRODUCTO ESPECIFICO)
#filtrar los datos del producto A

producto_a = ventas_productos[ventas_productos['Producto']== 'Producto A'] # hemos jalado el eje y 
fechas = producto_a['Fecha'] #hemos jalado el eje x 

plt.figure(figsize=(12,4)) # con esto le dimos tamaño a la figura
plt.plot(fechas , producto_a['Cantidad_Vendida'] , marker='o')
plt.show()
```

```python
#grafico circular (proporcion de ingresos por producto)

#los ingresos total por producto

ventas_productos['Ingresos'] = ventas_productos['Cantidad_Vendida'] * ventas_productos['Precio_Unitario']
ingresos_totales = ventas_productos.groupby('Producto')['Ingresos'].sum()

#crear el grafico circular 
plt.pie(ingresos_totales,labels = ingresos_totales.index , autopct='%1.1f%%')
plt.show()
```

```python
#Grafico de lineas Multiples (tmp media por mes para ambas ciudades)
ejex = temperaturas_ciudades['Mes']
datos_ciudad_x = temperaturas_ciudades
datos_ciudad_y = temperaturas_ciudades

```

```python
#grafico de dispersion MES VS TEMPERATURA MEDIA
#ciudad x
ciudad_x = temperaturas_ciudades[temperaturas_ciudades['Ciudad']== 'Ciudad X'] 

#CREAMOS EL GRAFICO DE DISPERSION
plt.scatter(ciudad_x['Mes'],ciudad_x['Temperatura_Media'])
plt.grid()
plt.show()

#hagan el de ciudad Y
```

```python
#grafico de barras de horizontales (poblacion de varios paises en un año especifico)

specific_year = 2020
poblacion_year = poblacion_paises[poblacion_paises['Año']==specific_year]

plt.barh(poblacion_year['País'], poblacion_year['Población'])
plt.title('Poblacion de paises por Año')
plt.xlabel('1M')
plt.ylabel('Pais')
plt.show()
```

```python
#Grafico de area (cambio de la poblacion a lo largo de un año para un pais)
pais_a = poblacion_paises[poblacion_paises['País']=='País A']

plt.fill_between(list(pais_a['Año']), pais_a['Población'] , alpha=0.2 , color ='green')
plt.plot(list(pais_a['Año']), pais_a['Población'] , marker='o')
plt.grid()
plt.show()
```
